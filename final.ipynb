```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
name=["Exp1","Exp2","Exp3","Exp4","Exp5"]
real=[[350,0],[350,-100],[450,-40],[550,-100],[550,-180]]

```

```python
index=0
for n in name:
    print(n)
    Y= pd.read_csv(n)
    X_mean=np.mean(Y['x'])
    Y_mean=np.mean(Y['y'])

    X_=[]
    Y_=[]
    X_real=[]
    Y_real=[]
    for t in Y['x']:
        X_.append(X_mean)
        Y_.append(Y_mean)
        X_real.append(real[index][0])
        Y_real.append(real[index][1])
    plt.figure(figsize=(24,12))
    plt.rc('axes', labelsize=30)
    plt.rc('axes',titlesize=30)
    plt.rc('legend', fontsize=30)
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.title("Эволюция компоненты X") # заголовок
    plt.xlabel("кадр") # ось абсцисс
    plt.ylabel("X, мм.") # ось ординат
    plt.grid()      # включение отображение сетки
    plt.plot(Y['t'], X_,'r',label="Среднее измеренное значение X")  # построение графика
    plt.plot(Y['t'], X_real,'g',label="Реальное значение X")  # построение графика
    plt.plot(Y['t'], Y['x'],'b',label="Измеренное значение X")  # построение графика
    plt.legend(loc=0)
    plt.savefig(n+' X')

    plt.figure(figsize=(24,12))
    plt.rc('axes', labelsize=30)
    plt.rc('axes',titlesize=30)
    plt.rc('legend', fontsize=30)
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.title("Эволюция компоненты Y") # заголовок
    plt.xlabel("кадр") # ось абсцисс
    plt.ylabel("Y, мм.") # ось ординат
    plt.grid()      # включение отображение сетки
    plt.plot(Y['t'], Y_,'r',label="Среднее измеренное значение Y")  # построение графика
    plt.plot(Y['t'], Y_real,'g',label="Реальное значение Y")  # построение графика
    plt.plot(Y['t'], Y['y'],'b',label="Измеренное значение Y")  # построение графика
    plt.legend(loc=0)
    plt.savefig(n+' Y')
    index+=1
```

```python
import math
x=["Exp1 [350,0]","Exp2 [350,-100]","Exp3 [450,-40]","Exp4 [550,-100]","Exp5 [550,-180]"]
X_parth=[]
Y_parth=[]
D_parth=[]
index=0
for n in name:
    print(n)
    Y= pd.read_csv(n)
    X_mean=np.mean(Y['x'])
    Y_mean=np.mean(Y['y'])
    X_parth.append((real[index][0]-X_mean))
    Y_parth.append((real[index][1]-Y_mean))
    data=(real[index][0]**2+real[index][1]**2)**0.5
    angle=math.atan2(real[index][1],real[index][0])
    D_parth.append(str(round(data))+" | "+str(round(angle*180/np.pi)))
    math.atan2(real[index][1],real[index][0])
    index+=1
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(24)    #  ширина Figure
fig.set_figheight(12)    #  высота Figure
ax.bar(x, X_parth)
plt.rc('axes', labelsize=18)
plt.rc('axes',titlesize=18)
plt.rc('legend', fontsize=18)
plt.xlabel("График абсолютной ошибки по X") # ось абсцисс
plt.ylabel("Абсолютная ошибка, мм") # ось ординат
plt.title("График ошибки расчета положения объекта относительно QR-кода") # заголовок
plt.savefig("График абсолютной ошибки по X")
```

```python
fig, ax = plt.subplots()
fig.set_figwidth(24)    #  ширина Figure
fig.set_figheight(12)    #  высота Figure
ax.bar(x, Y_parth)
plt.rc('axes', labelsize=18)
plt.rc('axes',titlesize=18)
plt.rc('legend', fontsize=18)
plt.xlabel("График абсолютной ошибки по Y") # ось абсцисс
plt.ylabel("Абсолютная ошибка, мм") # ось ординат
plt.title("График ошибки расчета положения объекта относительно QR-кода") # заголовок
plt.savefig("График абсолютной ошибки по Y")
```
